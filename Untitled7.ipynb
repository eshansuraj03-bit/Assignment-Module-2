```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sentences = [
    # Cricket (4)
    "The batsman scored a century in the match",
    "The bowler delivered a fast yorker",
    "The team won the cricket tournament",
    "Fielding plays a crucial role in cricket",

    # Cooking (3)
    "The chef prepared a delicious pasta dish",
    "Baking requires precise measurements of ingredients",
    "Spices enhance the flavor of food",

    # Cybersecurity (3)
    "Strong passwords help protect user accounts",
    "Encryption ensures secure communication",
    "Hackers exploit vulnerabilities in systems"
]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
```

```python
similarity_matrix = cosine_similarity(embeddings)

print("Similarity Matrix:")
print(similarity_matrix)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(similarity_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Sentence Similarity Heatmap")
plt.show()
```

```python
query = "The bowler took three wickets in one over"

query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, embeddings)[0]

# Get top 2 most similar sentences
top_indices = similarities.argsort()[-2:][::-1]

print("Query:", query)
print("\nTop 2 most similar sentences:\n")

for idx in top_indices:
    print(f"Sentence: {sentences[idx]}")
    print(f"Similarity Score: {similarities[idx]:.4f}\n")
```
