```python
import duckdb
import os
os.listdir()
con = duckdb.connect()
```

```python
#1
query1 = """
SELECT
    c.customer_id,
    c.customer_name,
    COUNT(o.order_id) AS total_orders
FROM read_csv_auto('customers (1).csv') c
LEFT JOIN read_json_auto('orders.json', format='array') o
ON c.customer_id = o.customer_id
GROUP BY c.customer_id, c.customer_name
ORDER BY total_orders DESC;
"""

con.execute(query1).fetchdf()
```

```python
#2
query2 = """
SELECT
    c.customer_name,
    SUM(o.total_amount) AS total_spent
FROM read_csv_auto('customers.csv') c
JOIN read_json_auto('orders.json') o
ON c.customer_id = o.customer_id
GROUP BY c.customer_name
ORDER BY total_spent DESC
LIMIT 3;
"""
con.execute(query2).fetchdf()
```

```python
#3
query3 = """
SELECT DISTINCT
    p.product_name
FROM read_csv_auto('customers.csv') c
JOIN read_json_auto('orders.json') o
ON c.customer_id = o.customer_id
JOIN read_parquet('products.parquet') p
ON o.order_id = p.order_id
WHERE c.customer_city = 'Bangalore';
"""

con.execute(query3).fetchdf()
```

```python
query4 = """
SELECT
    c.customer_name,
    o.order_date,
    p.product_name,
    p.quantity
FROM read_csv_auto('customers.csv') c
JOIN read_json_auto('orders.json') o
ON c.customer_id = o.customer_id
JOIN read_parquet('products.parquet') p
ON o.order_id = p.order_id;
"""

con.execute(query4).fetchdf()
```
